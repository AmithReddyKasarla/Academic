Question 3 (15 points). In this question, you are required to create visualization for time series data. 

We have collected all the articles from AMIA in the last 30 years. The dataset is avaliable on GitHub: https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv. 

Please use the tool Yake: https://github.com/LIAAD/yake to extract the top N (you can set different N value to get the best visualization results) keywords from the article titles (all the articles), then visualize how the keywords changed over years in one figure. (Split the data by every five years)

```python
# Your python code here:
#Importing all the required libraries and reading the data
!pip install yake
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yake as y
Df = pd.read_csv('https://raw.githubusercontent.com/unt-iialab/info5502-spring2022/main/datasets/lab_assignment_05_amia_cleaned.csv')
Df.head(2)





```

```python
#Declaring an empty list z
#In this I am taking N as 3, so the keywords we get for each year are 3
z = []
count = 1
#Using for loop for year in the loaded dataframe
for year in Df['year'].unique():
  language = "en"
  max_ngram_size = 3
  deduplication_thresold = 0.9
  deduplication_algo = 'seqm'
  windowSize = 1
  numOfKeywords = 3
  text = " ".join(title for title in Df[Df['year']==year]['title'].values)
  custom_kw_extractor = y.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)
  keywords = custom_kw_extractor.extract_keywords(text)

  for kw,v in keywords:
#Appending the keywords into the list z    
    z.append([kw,year])
  print(z)
    
```

```python
#Creating a new dataframe to store the keywords and year
new = pd.DataFrame(z,columns =['keywords', 'year'])
```

```python
#Importing plotly graph as gob
import plotly.graph_objects as gob
fig=gob.Figure(gob.Scatter(x=new.keywords, y=new.year))
fig['layout']['yaxis1'].update(title='Keywords Changed Over Years', range=[1980, 2025], dtick=5, autorange=False)
fig.update_layout(template = 'plotly_white')
fig.update_layout(xaxis = dict(ticktext=new.year))
fig.show()
    
```
